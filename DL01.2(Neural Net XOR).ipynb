# Create a simple neural network XOR

# Import required python libraries

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
import numpy as np
```

# Defining input and output

```python
inp = np.array([[0,0], [0,1], [1,0], [1,1]])
out = np.array([[0], [1], [1], [0]])
```

# Defining the Model

```python
model = Sequential()
model.add(Dense(4, input_shape=(2,)))
model.add(Activation('sigmoid'))
model.add(Dense(2, input_shape=(2,)))
model.add(Activation('sigmoid'))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

# Compiling the model

```python
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
model.summary()
```

# Training the model

```python
model.fit(inp, out, epochs=100)
```

```python
x_test = np.array([[0,0], [0,1], [1,0], [1,1]])
y_test = np.array([[0], [1], [1], [0]])
```

# Predicting the values

```python
print(f"Actual Values: {y_test}")
print(f"Predicted Values: {[[round(x)] for [x] in model.predict(x_test)]}")
```

# Evaluating the Model

```python
model_scores = model.evaluate(x_test, y_test)
```

```python
print(f"Accuracy: {model_scores[1] * 100}")
```
