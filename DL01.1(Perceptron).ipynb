# Create a perceptron and apply sigmoid activation

```python
import numpy as np
def sigmoid(inp):
  return 1/(1 + np.exp(-inp))
```

```python
def weighted_sum(values,weights):
  out = 0.0
  for i in range(len(values)):
    out+=values[i]*weights[i];
  return out
```

```python
var = int(input("Enter number of elements : "))
print("Enter the inputs")
input_values = []
for i in range(var):
    value = float(input())
    input_values.append(value)

input_weights = []
print("Enter the weights")
for i in range(var):
    weight = float(input())
    input_weights.append(weight)

print(f"Input Values : {input_values}")
print(f"Input Weights : {input_weights}")
print("The net input can be calculated as Yin = x1w1 + x2w2 + x3w3")
yin=weighted_sum(input_values,input_weights)
print(f"Yin = {yin}")

print(f"After applying Sigmoid : {sigmoid(yin)}")
```
